```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import re

%matplotlib inline
```

```python
temp = []
for fn in os.listdir('./data/memcpy'):
    numbers = re.findall(r'\d+', fn)
    print("Reading file " + fn + ". streams " + numbers[0] + " and chunks " + numbers[1])
    tempframe = pd.read_csv("./data/memcpy/"+fn, sep = ",")
    tempframe['streams']=numbers[0]
    temp.append(tempframe)
data_frame = pd.concat(temp, ignore_index = True)
```

```python
data_frame.columns = data_frame.columns.str.strip()
```

```python
data_frame[['streams']]=data_frame[['streams']].apply(pd.to_numeric)
```

```python
data_frame['totchunks']=data_frame.chunkno*data_frame.streams
```

```python
data_frame.columns
```

```python
data_frame['Single Stream Chunking nosync']=data_frame.chunk_nosync/data_frame.cudaMemcpy
data_frame['Stream Chunking']=data_frame.chunktimestream/data_frame.cudaMemcpy
data_frame['Stream Chunking nosync']=data_frame['chunkstreams_ nosync']/data_frame.cudaMemcpy
data_frame['Stream Chunk size']=data_frame['size']*data_frame['size']*8/data_frame['totchunks']
```

```python
data_frame.to_csv('./data/memcpy.csv',sep = ',', index=False)
```
